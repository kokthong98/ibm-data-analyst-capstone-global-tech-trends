### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Drop rows where DatabaseWantToWorkWith is missing
db_want = df['DatabaseWantToWorkWith'].dropna()

# Split the semi-colon separated string into lists
db_lists = db_want.str.split(';')

# Flatten the list of lists into one list of databases
all_dbs = [db.strip() for sublist in db_lists for db in sublist]

# Count frequency of each database
db_counts = Counter(all_dbs)

# Get the top 5 most common databases and their counts
top5 = db_counts.most_common(5)

# Separate names and counts for plotting
labels, sizes = zip(*top5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.axis('equal')  # Equal aspect ratio ensures pie chart is circular
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Drop missing values
dev_types = df['DevType'].dropna()

# Split the semicolon-separated roles into lists
dev_lists = dev_types.str.split(';')

# Flatten into one big list of developer roles
all_devs = [role.strip() for sublist in dev_lists for role in sublist]

# Count frequencies of each developer role
dev_counts = Counter(all_devs)

# Get top 5 roles and their counts
top5 = dev_counts.most_common(5)
labels, sizes = zip(*top5)

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Developer Roles')
plt.axis('equal')  # Keep pie chart circular
plt.show()
```
