### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Make sure Age is categorical with ordered age groups
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old', 
             '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']
df['Age'] = pd.Categorical(df['Age'], categories=age_order, ordered=True)

# Calculate median compensation by age group
median_comp_by_age = df.groupby('Age')['ConvertedCompYearly'].median()

# Plotting
plt.figure(figsize=(10,6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values, marker='o', linestyle='-', color='blue')
plt.title('Median Yearly Compensation by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation Yearly (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
import matplotlib.pyplot as plt
import pandas as pd

# Define full ordered age groups
age_order = ['Under 18 years old', '18-24 years old', '25-34 years old', 
             '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']

# Ensure Age is categorical and ordered
df['Age'] = pd.Categorical(df['Age'], categories=age_order, ordered=True)

# Filter for age groups between 25 and 45 inclusive
age_filter = ['25-34 years old', '35-44 years old', '45-54 years old']  # Include 45-54 for full coverage up to 45
df_filtered = df[df['Age'].isin(age_filter)]

# Calculate median compensation by filtered age groups
median_comp_filtered = df_filtered.groupby('Age')['ConvertedCompYearly'].median()

# Plotting
plt.figure(figsize=(8,5))
plt.plot(median_comp_filtered.index, median_comp_filtered.values, marker='o', linestyle='-', color='green')
plt.title('Median Yearly Compensation for Age Groups 25-45')
plt.xlabel('Age Group')
plt.ylabel('Median Converted Compensation Yearly (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# If 'YearsCodePro' exists and is numeric, use it directly.
# If it's categorical, convert to approximate numeric values:

def convert_years_code_pro(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    mapping = {
        'Less than 1 year': 0.5,
        'More than 50 years': 51,
        'Less than 1 year ': 0.5
    }
    if x in mapping:
        return mapping[x]
    try:
        return float(x)
    except:
        return np.nan

df['YearsCodePro_num'] = df['YearsCodePro'].apply(convert_years_code_pro)

# Drop rows with missing experience or job satisfaction
df_filtered = df.dropna(subset=['YearsCodePro_num', 'JobSatPoints_6'])

# Group by YearsCodePro_num (experience), compute median job satisfaction
job_sat_by_exp = df_filtered.groupby('YearsCodePro_num')['JobSatPoints_6'].median()

# Optional: Sort index to plot properly (experience ascending)
job_sat_by_exp = job_sat_by_exp.sort_index()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(job_sat_by_exp.index, job_sat_by_exp.values, marker='o', linestyle='-', color='purple')
plt.title('Median Job Satisfaction (JobSatPoints_6) by Years of Professional Coding Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

```python
import matplotlib.pyplot as plt

# Make sure 'Experience_num' and 'ConvertedCompYearly' columns exist and have no missing values
df_filtered = df.dropna(subset=['Experience_num', 'ConvertedCompYearly'])

# Group by experience and calculate median compensation
median_comp = df_filtered.groupby('Experience_num')['ConvertedCompYearly'].median().sort_index()

# Plotting
plt.figure(figsize=(10,6))
plt.plot(median_comp.index, median_comp.values, marker='o', color='blue')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Median Compensation (ConvertedCompYearly)')
plt.title('Median Compensation by Years of Professional Coding Experience')
plt.grid(True)
plt.show()
```
