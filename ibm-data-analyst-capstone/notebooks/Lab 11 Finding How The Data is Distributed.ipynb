### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
# Display column names
print("Columns:")
print(df.columns)

# Display data types of each column
print("\nData types:")
print(df.dtypes)

# Display summary info including number of non-null values and memory usage
print("\nDataFrame info:")
df.info()

# Optionally, get basic statistics for numeric columns
print("\nSummary statistics:")
print(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Count missing values in each column
missing_counts = df.isnull().sum()
print("Missing values per column:")
print(missing_counts[missing_counts > 0])  # Show only columns with missing values

df_clean = df.dropna()
print(f"New shape after dropping missing rows: {df_clean.shape}")

col = 'Employment'  # example column
mode_val = df[col].mode()[0]
df[col].fillna(mode_val, inplace=True)
print(f"Filled missing values in '{col}' with mode: {mode_val}")

col = 'ConvertedCompYearly'
median_val = df[col].median()
df[col].fillna(median_val, inplace=True)
print(f"Filled missing values in '{col}' with median: {median_val}")
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
# List of key columns
key_cols = ['Employment', 'JobSat', 'YearsCodePro']

# Loop through each and print value counts
for col in key_cols:
    print(f"Value counts for {col}:")
    print(df[col].value_counts(dropna=False))  # include NaN counts too
    print("\n" + "-"*40 + "\n")
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
import matplotlib.pyplot as plt

# Drop missing values for cleaner plot
jobsat_counts = df['JobSat'].value_counts()

plt.figure(figsize=(7,7))
jobsat_counts.plot.pie(autopct='%1.1f%%', startangle=140, counterclock=False)
plt.title('Job Satisfaction Distribution')
plt.ylabel('')  # Hide ylabel for cleaner look
plt.show()
```

```python
interpretation = """
The pie chart displays the distribution of job satisfaction scores among survey respondents.
The largest portions of respondents rated their satisfaction as 7.0 (21.9%) and 8.0 (25.8%), 
indicating that a majority of participants are generally satisfied with their jobs.
Lower satisfaction ratings (1.0 to 4.0) account for a small percentage of responses, 
showing that only a minority are dissatisfied.
Overall, the data suggests that job satisfaction tends to be on the positive side for most respondents.
"""
print(interpretation)
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code hereimport pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Assume df is your dataframe already loaded

# Split the languages (assuming ';' delimiter) and count frequencies
worked_langs = df['LanguageHaveWorkedWith'].dropna().str.split(';')
want_langs = df['LanguageWantToWorkWith'].dropna().str.split(';')

# Flatten the lists and count
worked_counts = Counter([lang.strip() for sublist in worked_langs for lang in sublist])
want_counts = Counter([lang.strip() for sublist in want_langs for lang in sublist])

# Create DataFrame for plotting
all_langs = set(worked_counts.keys()) | set(want_counts.keys())
data = {
    'Language': list(all_langs),
    'WorkedWith': [worked_counts.get(lang, 0) for lang in all_langs],
    'WantToWorkWith': [want_counts.get(lang, 0) for lang in all_langs]
}

lang_df = pd.DataFrame(data).sort_values(by='WorkedWith', ascending=False)

# For clarity, select top 10 languages by 'WorkedWith'
top_lang_df = lang_df.head(10)

# Plot grouped bar chart
fig, ax = plt.subplots(figsize=(12,6))
width = 0.35
x = range(len(top_lang_df))

ax.bar(x, top_lang_df['WorkedWith'], width=width, label='Have Worked With')
ax.bar([i + width for i in x], top_lang_df['WantToWorkWith'], width=width, label='Want To Work With')

ax.set_xticks([i + width / 2 for i in x])
ax.set_xticklabels(top_lang_df['Language'], rotation=45, ha='right')
ax.set_ylabel('Count')
ax.set_title('Comparison of Programming Languages: Worked With vs Want to Work With')
ax.legend()

plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data - count RemoteWork categories per Country
remote_by_country = df.groupby(['Country', 'RemoteWork']).size().reset_index(name='Count')

# For clarity, select top N countries by total respondents
top_countries = df['Country'].value_counts().head(10).index
remote_by_country_top = remote_by_country[remote_by_country['Country'].isin(top_countries)]

plt.figure(figsize=(14,7))
sns.barplot(data=remote_by_country_top, x='Country', y='Count', hue='RemoteWork')
plt.xticks(rotation=45, ha='right')
plt.title('Distribution of RemoteWork by Top 10 Countries')
plt.ylabel('Count of Respondents')
plt.xlabel('Country')
plt.legend(title='RemoteWork')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
!pip install scipy

def convert_years(x):
    if x == 'Less than 1 year':
        return 0
    elif x == 'More than 50 years':
        return 51
    try:
        return float(x)
    except:
        return None

df['YearsCodePro_num'] = df['YearsCodePro'].apply(convert_years)

df_corr = df.dropna(subset=['YearsCodePro_num', 'JobSat'])

pearson_corr = df_corr['YearsCodePro_num'].corr(df_corr['JobSat'], method='pearson')
spearman_corr = df_corr['YearsCodePro_num'].corr(df_corr['JobSat'], method='spearman')

print(f"Pearson correlation: {pearson_corr:.3f}")
print(f"Spearman correlation: {spearman_corr:.3f}")
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
import pandas as pd
import matplotlib.pyplot as plt

crosstab = pd.crosstab(df['EdLevel'], df['Employment'])
print(crosstab)

crosstab_norm = crosstab.div(crosstab.sum(axis=1), axis=0)

crosstab_norm.plot(kind='bar', stacked=True, figsize=(12,7))
plt.title('Employment Status Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Proportion')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
